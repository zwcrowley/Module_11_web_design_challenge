## Requirement 1: Create Plots to Showcase the Relationship Between Weather Variables and Latitude

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
# Read saved data
city_data_df = pd.read_csv("Resources/cities.csv", index_col="City_ID")

# Display sample data
city_data_df.head()
```

### Create the Scatter Plots Requested

#### Latitude Vs. Temperature

```python
# Build scatter plot for latitude vs. temperature
# Use seaborn plot to make the scatter plot, add a title and axis labels:
sns.set_theme(style="darkgrid", font_scale=1.25, rc={'xtick.bottom': True,
    'ytick.left': True,})
# Set fig and axis, and figure size:
fig, ax = plt.subplots(figsize=(8, 5))
# Create the plot, take out conf int, set color and ax:
g = sns.scatterplot(ax=ax, data=city_data_df, x="Lat", y="Max Temp", s=20, color="darkblue")
# city_data_df.plot(x='Lat', y='Max Temp', kind = "scatter", edgecolor='k', s=40)
g.set_xlim(-80, 100)
g.set_ylim(-60, 60)

# Incorporate the other graph properties
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (C)")
plt.title("City Latitude vs Max Temperature (11-4-2022)")

# Save the figure:
plt.savefig("assets/images/Fig1.png")

# Show plot:
plt.show()
```

#### Latitude Vs. Humidity

```python
# Build the scatter plots for latitude vs. humidity
# Use seaborn plot to make the scatter plot, add a title and axis labels:
sns.set_theme(style="darkgrid", font_scale=1.25, rc={'xtick.bottom': True,
    'ytick.left': True,})
# Set fig and axis, and figure size:
fig, ax = plt.subplots(figsize=(8, 5))
# Create the plot, take out conf int, set color and ax:
g = sns.scatterplot(ax=ax, data=city_data_df, x="Lat", y="Humidity", s=20, color="darkblue")

g.set_xlim(-80, 100)
g.set_ylim(-20, 120)

# Incorporate the other graph properties
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.title("City Latitude vs Humidity (11-4-2022)")

# Save the figure:
plt.savefig("assets/images/Fig2.png")

# Show plot:
plt.show()
```
